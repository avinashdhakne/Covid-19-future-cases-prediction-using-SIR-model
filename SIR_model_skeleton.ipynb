```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ndays = 100
dt = 1
beta = 1.0/3.0
gamma = 1.0/14.0
```

```python
S = np.zeros(ndays)
I = np.zeros(ndays)
R = np.zeros(ndays)
t = np.arange(ndays)*dt
```

```python
I[0] = 0.001
S[0] = 1 - I[0]
R[0] = 0.0
```

```python
for i in range(ndays-1):
  S[i+1] = S[i] - beta*(S[i]*I[i])*dt
  I[i+1] = I[i] + (beta*S[i]*I[i] - gamma*I[i]) *dt
  R[i+1] = R[i] +(gamma*I[i])*dt
```

```python
plt.plot(t,S)
plt.plot(t,I)
plt.plot(t,R)
plt.savefig('foo.png')
```
